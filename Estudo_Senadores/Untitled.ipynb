# Análise do Eleitorado brasileiro
 

```python
# importando as bilbiotecas
import pandas as pd
```

```python
# carregando o arquivo CSV
df = pd.read_csv('eleitorado_municipio_2020.csv', encoding='latin1', sep=';')

df.head()
```

```python
df.head(3).T
```

```python
# Tipos e quantidade dos dados
df.info()
```

```python
# as cinco cidades com maior de eleitores deficiente
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitora:',df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitora:',df['QTD_ELEITORES_MASCULINO'].sum())
print('Não informado:',df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Criar variáveis para calcular percentual
tot_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao  = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
# Mostrar valor percentualmente
print('Eleitora:', ((tot_fem / tot_eleitores)*100).round(2))
print('Eleitores:', ((tot_masc / tot_eleitores)*100).round(2))
print('Não informado:', ((tot_nao / tot_eleitores)*100).round(2))
```

```python
# Quantas cidade com mais homens que mulheres?
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Criar uma nova coluna para indicar a relação fem/masc
df['RELACAO_FEM/MSC'] = df['QTD_ELEITORES_MASCULINO'] / df['QTD_ELEITORES_FEMININO']
```

```python
# Quais são os municípios com maior relação f/m
df.nlargest(5,'RELACAO_FEM/MSC').T
```

```python
Quais são os municípios com menor relação f/m

df.nsmallest(5, 'RELACAO_FEM/MSC').T
```

```python
# Vamos criar um novo dataframe só com municípios do Brasil

df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head(3).T
```

```python
df_brasil.nlargest(5,'RELACAO_FEM/MSC').T
```

```python
%matplotlib inline


# plotar um gráfico de distribuição da relação f/m

df_brasil['RELACAO_FEM/MSC'].plot.hist(bins=100)
```

```python
# importar as bibliotecas de grafico seaborn

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plotar a distribuição  da relação F/M

sns.distplot(df_brasil['RELACAO_FEM/MSC'], bins=100, color='red', kde=False)

# deixando o gráfico mais bonito
plt.title('Relação Eleitoras/Eleitores', fontsize=18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Freqeência', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```
